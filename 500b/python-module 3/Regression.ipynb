```python
#ordinary least squares reegression
import statsmodels.api as sm 
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#col1: country, col2: ease of access for generic drugs, private sector, col3: col2, but public
data = pd.read_csv('ex3_mod_2.csv', header=0)
data
```

```python
plt.scatter(data.private,data.public)
lr_model = sm.OLS(data.private,data.public).fit()
print(lr_model.summary())
```
